숫자를 건넬 때 일부 자릿수를 영단어로 바꾼 카드를 건네주면 프로도는 원래 숫자를 찾는 게임입니다.

다음은 숫자의 일부 자릿수를 영단어로 바꾸는 예시입니다.

1478 → "one4seveneight"  
234567 → "23four5six7"  
10203 → "1zerotwozero3"    

이렇게 숫자의 일부 자릿수가 영단어로 바뀌어졌거나,   
혹은 바뀌지 않고 그대로인 문자열 s가 주어졌을때
s가 의미하는 원래 숫자를 출력하도록 코드를 작성해주세요

s = "77three4one

---
참고로 각 숫자에 대응되는 영단어는 다음 표와 같습니다.

숫자	영단어  
0	zero  
1	one  
2	two  
3	three  
4	four  
5	five  
6	six  
7	seven  
8	eight  
9	nine  

```python
## 나의 틀린 오답 !!! ##
# keys = list(range(0,10))
# values = ['zero','one','two','three','four','five','six','seven','eight','nine']
# dict_num = dict(zip(keys,values))

# i = 0
# s = "77three4one"
# while i < len(s):
#     if type(s[i]) == str:      ## 수정할 점(1) s = s[i] = 문자열 !!
#         print(dict_num[s[i]])  ## 수정할 점(2) dic key는 int/ 접근값 = str
#     else:
#         print(s[i])
#     i += 1

## (1) s가 문자열, s[i]도 무조건 문자열. 즉 무조건 True.
#####  숫자냐, 영단어냐 구분은 type()가 아니라 문자열 내용으로 구분했어야,,,

## (2) dic의 key에서 7은 int고, 접근하는 값 "7" 은 str
#####  keyerror!! 어쨌든 문제는 영단어 -> 숫자 바꾸는 문제이므로
#####  key와 value를 바꿨어야 함.
```

```python
# 오답 보완하기
word_to_num = {
    "zero": "0",
    "one": "1",
    "two": "2",
    "three": "3",
    "four": "4",
    "five": "5",
    "six": "6",
    "seven": "7",
    "eight": "8",
    "nine": "9",
}
s = "77three4one"

temp = ""      # 영단어 모을 임시 문자열
result = ""    # 최종 숫자 문자열

for ch in s:   # 메인루프: 문자 하나씩 훑기
    if ch.isdigit():   # ch.isdigit() -> ch가 문자열일 때 사용가능한 문자열 메서드.
        result += ch   # 이 문자열이 숫자문자로만 되있는지 True/False
    else:
        temp += ch     # t -> th -> thr -> ...
        if temp in word_to_num:  # temp에서 완성된 단어가 key값이면
            result += word_to_num[temp]   # 숫자로 바꿔 결과 추가
            temp = ""
print(result)
```

```python
## 답안
for key in word_to_num:
    s = s.replace(key,word_to_num[key])

print(s)

## '원본'.replace('바꿀거', '이걸로', [최대횟수]) <- "문자열 메서드"
## '원본'문자열 안에서 '바꿀거'를 찾아 '이걸로' 바꾼 새 문자열을 돌려주는 메서드
## [최대횟수]를 쓰지 않으면 전부 다 바꿈.
## s = s.replace(key,word_to_num[key])처럼 값을 할당해야 진짜 바꿈 !!!
```
