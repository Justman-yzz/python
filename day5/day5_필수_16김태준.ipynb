__[문제6]__

5번 문제에서 생성한 인스턴스의 이름, 나이, 성별을 출력해주세요.



출력 결과  ) <br>
이름 : 김코  
나이 : 30

__Tips__


- 인스턴스 변수에 접근하여 값을 출력하면 됩니다  : )

```python
class 오즈:
    def __init__(self, name, age, coding_level):
        self.name = name
        self.age = age
        self.coding_level = coding_level

coding = 오즈("김코",30,"A")

print("이름:",coding.name)
print("나이:",coding.age)
print("코딩레벨:",coding.coding_level)
```

__[문제7]__

오즈 (oz) 클래스에서 이름, 나이, 성별을 출력하는 who() 메소드를 추가하고 who()메소드를 이용해 아래 출력 결과를 만들어주세요


출력 결과 ) <br>
이름: 김코, 나이: 30, 코딩레벨: A

```python
class 오즈:
    def __init__(self, name, age, coding_level):
        self.name = name
        self.age = age
        self.coding_level = coding_level

    def who(self):
        print(f"이름: {self.name}, 나이: {self.age}, 코딩레벨: {self.coding_level}")

coding = 오즈("김코",30,"A")
coding.who()

# print() 대신 return 을 사용.
```
