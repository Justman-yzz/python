## 모험가들의 레벨업 도전

### [배경 이야기]

당신은 마법사, 전사, 도적, 힐러 네 명의 모험가를 이끄는 리더입니다.

각 캐릭터는 힘, 민첩, 지능, 체력이라는 능력치를 가지고 있습니다. 이 능력치들을 이용해 다양한 도전을 해결하고 사냥을 통해 경험치를 얻어 레벨업을 해야 합니다.

<br>

### [동료 캐릭터와 능력치 정보]

마법사: 힘(10), 민첩(20), 지능(80), 체력(50)

전사: 힘(80), 민첩(30), 지능(20), 체력(100)

도적: 힘(40), 민첩(80), 지능(30), 체력(60)

힐러: 힘(20), 민첩(40), 지능(70), 체력(70)

<br>

* 변수명을 영어로 변경해서 풀어보세요 !

```python
# 마법사 능력치 정보
wizard__str = 10
wizard__dex = 20
wizard__int = 80
wizard__hp = 50

# 전사 능력치 정보
warrior__str = 80
warrior__dex = 30
warrior__int = 20
warrior__hp = 100

# 도적 능력치 정보
thief__str = 40
thief__dex = 80
thief__int = 30
thief__hp = 60

# 힐러 능력치 정보
healer__str = 20
healer__dex = 40
healer__int = 70
healer__hp = 70


# "능력치 또는 경험치 확인" "메시지" 리스트
text__condition = ("이 아니므로","이므로")

# 능력치에 따른 "스킬사용 가능 여부" "메시지" 리스트
text__skill = ("할 수 없습니다", "할 수 있습니다")
```

<br>

__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다.

마법을 사용하려면 지능이 50 이상이어야 합니다.

마법사의 지능이 50 이상인지 확인하세요.

```python
skill__wizard = ( wizard__int >= 50 )
question__wizard = text__condition[skill__wizard]
skill__available__wizard = text__skill[skill__wizard]

print(f'마법사는 지능이 50 이상{question__wizard} 마법을 사용{skill__available__wizard}.')
```

<br>

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다.

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
skill__warrior = ( warrior__str >= 60 ) and ( warrior__hp >= 80 )
question__warrior = text__condition[skill__warrior]
skill__available__warrior = text__skill[skill__warrior]

print(f'전사는 힘이 60 이상, 그리고 체력이 80 이상{question__warrior} 강력한 공격을 {skill__available__warrior}.')
```

<br>

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다.

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
skill__thief = ( thief__dex >= 70 ) or ( thief__str >= 50 )
question__thief = text__condition[skill__thief]
skill__available__thief = text__skill[skill__thief]

print(f'도적은 민첩이 70 이상, 또는 힘이 50 이상{question__thief} 몬스터를 회피하고 공격을 {skill__available__thief}.')
```

<br>

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다.

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
skill__healer = ( healer__int >= 60 ) and ( healer__hp >= 50 )
question__healer = text__condition[skill__healer]
skill__available__healer = text__skill[skill__healer]

print(f'힐러는 지능이 60 이상, 그리고 체력이 50 이상{question__healer} 파티의 체력을 회복{skill__available__healer}.')
```

<br>

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다.


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
# "경험치"에 따른 "레벨업 메시지" 리스트
text__exp = ("하려면 조금 더 힘내셔야 합니다 ㅠㅠ", " !!! ")

# 모든 캐릭터의 경험치 정보
exp__wizard = 120
exp__warrior = 150
exp__thief = 110
exp__healer = 90

# 동료 캐릭터 (0 = 마법사, 1 = 전사, 2 = 도적, 3 = 힐러 )
charater = ("마법사", "전사", "도적", "힐러")

# 레벨업 조건 (0 = 마법사, 1 = 전사, 2 = 도적, 3 = 힐러 )
level_up = (exp__wizard >= 100, exp__warrior >= 100, exp__thief >= 100, exp__healer >= 100)


print(f'{charater[0]}는 경험치가 100{text__condition[level_up[0]]} 레벨업{text__exp[level_up[0]]}')
print(f'{charater[1]}는 경험치가 100{text__condition[level_up[1]]} 레벨업{text__exp[level_up[1]]}')
print(f'{charater[2]}는 경험치가 100{text__condition[level_up[2]]} 레벨업{text__exp[level_up[2]]}')
print(f'{charater[3]}는 경험치가 100{text__condition[level_up[3]]} 레벨업{text__exp[level_up[3]]}')
```
