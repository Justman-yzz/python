````python
list_user = []

while True:
    변수 = input("숫자만 입력해주세요. 끝내고 싶으면 done을 입력: ")
    
    if 변수 == "done":
        print("입력된 숫자:", list_user)
        break
    try:
        num = int(변수)
        list_user.append(num)
    except ValueError:
        print(f"잘못된 값: {변수}. 숫자를 입력해주세요.")

````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time

def time_decorator(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f'함수 실행 시간:{(end - start):.2f}초')
        return result
    return wrapper

@time_decorator
def arbitary_function(a,b):
    a ** b

arbitary_function(2,100)
````
